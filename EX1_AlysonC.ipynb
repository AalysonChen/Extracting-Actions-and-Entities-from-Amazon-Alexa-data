```
import spacy 
from collections import Counter
nlp = spacy.load('en')
```

```
file_name= ('drive/My Drive/Colab Notebooks/cleaned.alexadata.csv')
file_text= open(file_name).read()
doc=nlp(file_text)
#Extract tokens for the given doc
print ([token.text for token in doc ])
```

```
#Find the Top Five Action Words
words = [token.text for token in doc if token.is_stop != True and token.is_punct != True]
word_freq = Counter(words)
common_words = word_freq.most_common(6) #the top one action word is \n (newline) so Instead of finding word_freq.most_common(5), 
                                        #I tried to use word_freq.most_common(6) to find the top five action words.(ingored '\n')
print(common_words)
```

**1. Summary for Top Five Action Words**:

After analyzing the dataset, I found the top five action words and they're 'Alexa', 'play', 'timer', 'volume','pause' respectively. The result reflected the top five popular ways to use Alexa. According to the report from sayspring.com, it stated that set a timer, play a song and hears the news are the most attractive features. As result, the outcome of this text analytics seems to be convincing.

```
# Find the Top Five Categories 
# Step1: Name Entities
entities=[(i,i.label_, i.label) for i in doc.ents]
entities
```

```
#Step 2: find the top five highest categories 
from heapq import nlargest
ents= Counter(entities) 
high = ents.most_common(5)

print(" Top Five Highest Categories:") 
print("Keys: Values")
for i in high: 
    print(i[0]," "," ") 
```

```
#Just play around to find the most frequent category 
def most_frequent(entities): 
    return max(set(entities), key = entities.count) 
  
entities=[(i,i.label_, i.label) for i in doc.ents]
print(most_frequent(entities)) 
```

**2. Summary for Top Five Categories:**

As we can see above, the top five categories of Alexa dataset are all related to the time setting function of Alexa. Therefore, we can assume that most of their customers used Alexa to set a timer, wake them up or schedule a date for meeting for customers.
